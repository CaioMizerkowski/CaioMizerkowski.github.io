```python
import seaborn as sns
from matplotlib import pyplot as plt
import polars as pl
```

```python
vitamin_trials = pl.read_csv('../datasets/P15-VitaminTrials.csv')
vitamin_trials
```

```python
plt.figure(figsize=(15, 15))
plt.title('Sem classificação')
sns.scatterplot(data=vitamin_trials.to_pandas(), x='Control Test Score', y='Vitamin Dose MG')
plt.show()
plt.close()
```

### Cores aleatórias
Ao aplicarmos um conjunto de cores sem considerar a paleta usada, chegamos a resultados que não evidenciam os padrões presentes nos dados.

```python
plt.figure(figsize=(15, 15))
plt.title('Cores Aleatórias')
hue_order = ['Children 8-12',
 'Children 12-14',
 'Children 15-18',
 'Men 18-24',
 'Men 25-34',
 'Men 35-60',
 'Women 18-24',
 'Women 25-34',
 'Women 35-60']
sns.scatterplot(data=vitamin_trials.to_pandas(), y='Control Test Score', x='Vitamin Dose MG', hue='Group', hue_order=hue_order)
plt.show()
plt.close()
```

### Tríade de cores
Se agruparmos os dados, reduzindo as categórias, e aplicarmos um conjunto de cores que respeitem a diferença de 60 graus entre elas no circulo de cores, maximizamos o contrantes entre os grupos e com isso se tornam vísiveis padrões antes não conhecidos. Como por exemplo o presente no grupo 'Children'.

```python
plt.figure(figsize=(15, 15))
plt.title('Tríade de cores')
vt_tmp = vitamin_trials.clone()
vt_tmp['Group'] = vt_tmp['Group'].apply(lambda x: x.split(' ')[0])
palette = ['#7316F6', '#77FC00', '#FF9000']
hue_order=['Children','Men','Women']
sns.scatterplot(data=vt_tmp.to_pandas(), y='Control Test Score', x='Vitamin Dose MG', hue='Group', palette=palette, hue_order=hue_order)
plt.show()
plt.close()
```

### Complementares divididas
Quando um grupo possui maior importância em nossas analises que outros dois, podemos usar as cores complementares divididas. Isso maximizará o contraste entre uma categoria e o todo. No nosso exemplo, isso tornará o grupo 'Children' mais visível.  

```python
plt.figure(figsize=(15, 15))
plt.title('Complementares Divididas')
vt_tmp = vitamin_trials.clone()
vt_tmp['Group'] = vt_tmp['Group'].apply(lambda x: x.split(' ')[0])
palette = ['#8105D0', '#CCFA00', '#FFE100']
hue_order=['Children','Men','Women']
sns.scatterplot(data=vt_tmp.to_pandas(), y='Control Test Score', x='Vitamin Dose MG', hue='Group', palette=palette, hue_order=hue_order)
plt.show()
plt.close()
```

### Monocromático
Podemos levar as categorias menos relevates para a escala monocromática, aumentando ainda mais o contraste.

```python
plt.figure(figsize=(15, 15))
plt.title('Triade de cores')
vt_tmp = vitamin_trials.clone()
vt_tmp['Group'] = vt_tmp['Group'].apply(lambda x: x.split(' ')[0])
palette=['#8105D0', '#ddd','#aaa']
hue_order=['Children','Men','Women']
sns.scatterplot(data=vt_tmp.to_pandas(), y='Control Test Score', x='Vitamin Dose MG', hue='Group', palette=palette, hue_order=hue_order)
plt.show()
plt.close()
```

### Cores análogas
Por último, podemos usar cores analógas para separar um dos grupos e trazer visualmente a informação que estas categorias são sequênciais.

```python
plt.figure(figsize=(15, 15))
plt.title('Cores análogas')
vt_tmp = vitamin_trials.clone()
vt_tmp['Group'] = vt_tmp['Group'].apply(lambda x: x.split(' ')[0] if 'Children' not in x else x)
palette=['#3C0AD3', '#6807D1', '#A203CF', '#ddd','#aaa']
hue_order=['Children 8-12','Children 12-14','Children 15-18','Men','Women']
sns.scatterplot(data=vt_tmp.to_pandas(), y='Control Test Score', x='Vitamin Dose MG', hue='Group', palette=palette, hue_order=hue_order)
plt.show()
plt.close()
```
